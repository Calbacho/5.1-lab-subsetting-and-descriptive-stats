# Challenge 1

#### In this challenge we will use the `Temp_States.csv` file. 

#### First import it into a data frame called `temp`.

```python
temp=pd.read_csv('Temp_States.csv')
```

#### Print `temp`

```python
print(temp)
```

#### Explore the data types of the Temp dataframe. What type of data do we have? Comment your result.

```python
temp.info()
# We have the the temperature for each city and its state.
```

#### Select the rows where state is New York

```python
temp[temp.State == 'New York']
```

#### What is the average of the temperature of cities in New York?

```python
temp[temp.State == 'New York'].Temperature.mean()
```

#### We want to know cities and states with Temperature above 15 degress Celcius

```python
temp[['City', 'State']][temp.Temperature > 15]
```

#### Now, return only the cities that have a temperature above 15 degress Celcius

```python
temp.City[temp.Temperature > 15]
```

#### We want to know which cities have a temperature above 15 degrees Celcius and below 20 degrees Celcius

*Hint: First write the condition then select the rows.*

```python
temp.City[(temp.Temperature > 15) & (temp.Temperature < 20)]
```

#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
temp[['State', 'Temperature']].groupby(temp.State).mean()
```

# Challenge 3

#### Open the Orders.csv dataset. Name your dataset orders

```python
import zipfile

zf = zipfile.ZipFile('Orders.zip')
orders = pd.read_csv(zf.open('Orders.csv'))
```

#### Explore your dataset by looking at the data types and the summary statistics. Comment your results

```python
orders.info()
```

```python
orders.describe().T
```

####  What is the average Purchase Price?

```python
orders.amount_spent.mean()
```

#### What were the highest and lowest purchase prices? 

```python
orders.amount_spent.max()
```

```python
orders.amount_spent.min()
```

#### Select all the customers we have in Spain

```python
orders.CustomerID[orders.Country == 'Spain']
```

#### How many customers do we have in Spain?
##### Hint : Use value_counts()

```python
orders.Country.value_counts()['Spain']
```

#### Select all the customers who have bought more than 50 items ?

```python
orders.groupby('CustomerID').sum()[orders.Quantity > 50].index
```

#### Select orders from Spain that are above 50 items

```python
orders[(orders.Quantity > 50) & (orders.Country == 'Spain')]
```

#### Select all free orders

```python
orders[orders.amount_spent == 0]
```

#### Select all orders that are 'lunch bag'
#### Hint: Use string functions

```python
orders[orders.Description.str.contains('lunch bag')]
```

#### Select all orders that are made in 2011 and are 'lunch bag'  

```python
orders[(orders.Description.str.contains('lunch bag')) & (orders.year == 2011)]
```

#### Show the frequency distribution of the amount spent in Spain.

```python
sns.distplot(orders.amount_spent[orders.Country == 'Spain'])
```

#### Select all orders made in the month of August

```python
orders[orders.month == 10]
```

#### Select how many orders are made by countries in the month of August
##### Hint: Use value_counts()

```python
orders.Country[orders.month == 10].value_counts()
```

#### What's the  average amount of money spent by country

```python
orders.groupby('Country').mean().amount_spent
```

#### What's the most expensive item?

```python
orders.Description[orders.UnitPrice == orders.UnitPrice.max()]
```

#### What was the average amount spent per year ?

```python
orders.groupby('year').mean().amount_spent
```
